Importing required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
Art=Image.open("Art.jpg")
Art
```

```python
type(Art) 
```

```python
Art_arr = np.asarray(Art) # Image object to ndarray conversion
Art_arr
```

```python
plt.imshow(Art_arr) 
```

```python
Art=Art_arr>>4 # Right Shift 4 bits
Art=Art<<4     # Left Shift 4 bits to clear 4 LSBs 
plt.imshow(Art)
```

```python
Art.shape #shape check
```

```python
secret=Image.open("image.png")
secret=np.asarray(secret)
plt.imshow(secret)
```

```python
secret.shape # png images have 4 channels
```

```python
secret = cv2.cvtColor(secret, cv2.COLOR_BGRA2BGR) # Converting 4 channel to 3 channel image to match no of channels with Art jpeg image
secret.shape
```

```python
secret=secret>>4        # Setting 4 MSBs of the Image to zero
plt.imshow(secret)
```

```python
Art_XOR_Secret=(Art ^ secret)  # Xor Art and secret
plt.imshow(Art_XOR_Secret)
```

```python
plt.imshow(Art_XOR_Secret<<4)   # Decoding Secret from Art
```

```python
encoded_image = Image.fromarray(Art_XOR_Secret) # Saving The Results
encoded_image.save("encoded_image.jpg")
decoded_image = Image.fromarray(Art_XOR_Secret<<4)
decoded_image.save("decoded_image.jpg")
shifted_image = Image.fromarray(secret)
shifted_image.save("shifted_image.jpg")
```
